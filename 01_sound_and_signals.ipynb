# 1. Sound and Signals

Create Cosine and Sine waves of different frequencies and mix them together. Extract a waveform worth 0.5sec duration and plot it.

```python
import thinkdsp

cos_sig = thinkdsp.CosSignal(freq=440, amp=1.0, offset=0)
sin_sig = thinkdsp.SinSignal(freq=880, amp=0.5, offset=0)

mix = sin_sig + cos_sig

wave = mix.make_wave(duration=0.5, start=0, framerate=11025)

wave.plot()
```

Extract the period of the mixed waveform, extract a section of the waveform with duration thrice that of its period. Plot the waveform.

```python
period = mix.period
segment = wave.segment(start=0, duration=period*3)
segment.plot()
```
