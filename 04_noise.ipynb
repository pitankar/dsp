# Uncorrelated Noise

More specifically, "uniform uncorrelated unifor noise", uniform meaning that the signal contains random values from uniform distribution and that every value in the range is equally likely. "Uncorrelated" would mean values are independent of each other, knowing one gives no information about the other.

```python
import thinkdsp
```

```python
signal = thinkdsp.UncorrelatedUniformNoise()
wave = signal.make_wave(duration=0.05, framerate=11025)

wave.plot()
```

```python
spectrum = wave.make_spectrum()
spectrum.plot_power()
```

# Integrated Spectrum

```python
integ = spectrum.make_integrated_spectrum()
integ.plot_power()
```

# Brownian Noise

```python
signal = thinkdsp.BrownianNoise()
wave = signal.make_wave(duration=0.5, framerate=11025)
wave.plot()
```

```python
import thinkplot
spectrum = wave.make_spectrum()
spectrum.plot_power(linewidth=1, alpha=0.5)
thinkplot.config(xscale='log', yscale='log')
```

# Gaussian Noise

```python
import thinkstats2
```

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
wave = signal.make_wave(duration=0.5, framerate=11025)
spectrum = wave.make_spectrum()

thinkstats2.NormalProbablityPlot(spectrum.real)
thinkstats2.NormalProbablityPlot(spectrum.imag)
```
