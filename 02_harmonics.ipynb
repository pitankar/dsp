# 2. Harmonics 

## Triangle Wave

```python
import thinkdsp

signal = thinkdsp.TriangleSignal(200)

wave = signal.make_wave(duration=0.5, framerate=10000)

period = signal.period
seg = wave.segment(start=0, duration=period*3)
seg.plot()
```

```python
spectrum = wave.make_spectrum()
spectrum.plot()
```

## Square Wave

```python
sq_signal = thinkdsp.SquareSignal(100)
sq_wave = sq_signal.make_wave(duration=0.5, framerate=10000)

period = sq_signal.period
sq_seg = sq_wave.segment(start=0, duration=period*3)
sq_seg.plot()
```

```python
spectrum = sq_wave.make_spectrum()
spectrum.plot()
```

## Filtering

```python
s = sq_wave.make_spectrum()
s.low_pass(1000)
s.plot()
```

```python
s = sq_wave.make_spectrum()
s.high_pass(500)
s.plot()
```
